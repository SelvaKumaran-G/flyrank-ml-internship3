## 2. The question: decision, action, cost of a wrong call

*What decision does your work improve? Who acts on it? What does a wrong recommendation cost?*

## The question

**Decision:** Which pages should be prioritized for content refresh this month,
given limited writer/editor time?

**Unit of analysis:** One page (identified by page_id, belonging to one client).

**Action:** An SEO strategist adds the top-ranked pages from my output to the
next sprint's content refresh queue, instead of relying on gut feel or a single
metric like "traffic dropped."

**Cost of a wrong call:**
- False positive (flagged as declining, but it isn't really): wastes writer
  hours refreshing a page that didn't need it, and a genuinely declining page
  goes untouched instead.
- False negative (missed a real decline): the page keeps losing traffic/rankings
  unnoticed until it's a bigger problem to fix.

Because both errors have a real cost, the model needs to be meaningfully better
than random or a naive rule — not just directionally better.

## 3. Quick look at the data (2-3 real numbers)

*Load the starter CSV below and show 2-3 real numbers that make your lane look worth the next 7 weeks.*

```python
import pandas as pd

df = pd.read_csv('/content/content_refresh_anonymized.csv')

# Number 1: how common is decline?
declining_share = (df['trend_direction'] == 'down').mean()
print(f"Share of pages currently declining: {declining_share:.1%}")

# Number 2: how stale is content on average?
if 'days_since_last_update' in df.columns:
    avg_staleness = df['days_since_last_update'].mean()
    print(f"Average days since last update: {avg_staleness:.0f}")

# Number 3: how much traffic is at stake?
avg_impressions = df['impressions_90d'].mean()
print(f"Average impressions (90d) per page: {avg_impressions:,.0f}")
```

Roughly [X]% of pages in this sample are declining, and pages carry an average
of [Y] impressions over 90 days — meaning even a modest improvement in
prioritization touches meaningful traffic. Combined with content going an
average of [Z] days between updates, there's a real backlog problem worth
triaging with data rather than a flat "refresh everything" or "refresh nothing"
approach.

## 4. Careful words: what I can and can't claim

*Write what your work will be able to say (observed, directional, decision-support) — and what it never will (causal proof, 'predicting Google').*

## What I can and can't claim

**What I can say:** This model produces an *observed, directional* risk score —
pages that look statistically similar to past pages that declined. It's
*decision-support*, meant to help a human prioritize, not replace judgment.

**What I can't say:** This is not causal proof that any specific factor *causes*
decline — it's correlation learned from historical patterns. It is not
"predicting Google's algorithm" or guaranteeing any individual page's future.
It also only reflects the client base and time window in this sample, so it may
not generalize to very different sites or industries.
