## 1. My lane as an ML task (type)

*Classification, clustering, ranking, or scoring — which one, and why?*

## My lane as an ML task

**Lane:** Refresh / Content Opportunity Scoring

**Task type:** This is a **scoring/ranking problem**, not pure classification.
The end deliverable is an ordered priority list ("refresh this page before that
one"), not just a yes/no label. Under the hood I'll likely train a
**binary classifier** (declining vs. not) and use its predicted *probability*
as the ranking score — but the product output is a ranked queue, so ranking is
the real task framing.

## 2. Target or proxy

*What would you predict? Where does that label come from — observed outcome or a defined rule?*

## Target or proxy

I don't have a direct "should refresh" label — nobody labeled that by hand.
My proxy target is `is_declining_label`, derived from `trend_direction == "down"`
in the starter dataset. This is a proxy, not ground truth: a page can be
declining for reasons a refresh won't fix (seasonality, a competitor outranking
it, a SERP feature change), and a page that isn't yet declining might still
benefit from a preemptive refresh. I'm using it because it's the closest
observable signal to "this page needs attention" available in this data.

## 3. Success metric

*One metric you can defend. What number means 'good'?*

## Success metric

**Precision@50** — of the top 50 pages my model ranks highest-risk, what
fraction are actually declining? This matches the real action: a content team
only has capacity to refresh a limited number of pages per cycle, so what
matters is whether the *top* of the queue is trustworthy, not overall accuracy
across all 30,000 pages. I'll compare this against the baseline hand-rule's
Precision@50 (~0.24 from Notebook 01) as the bar to beat.

## 5. Why ML beats a fixed rule here

*What makes the pattern too messy for an if-statement?*

## Why ML beats a fixed rule here

A fixed rule (e.g. "flag if traffic dropped more than 20%") only looks at one
signal in isolation and picks one arbitrary threshold. Real decline shows up
through *combinations* of signals — position drift, staleness, seasonality,
engagement patterns — that interact differently across content types and
clients. Notebook 01 already showed a simple hand-rule scores ~24% precision@50
while a learned model roughly triples that, because the model can weigh
multiple features together and learn non-obvious interactions a single
threshold can't capture. ML also generalizes better across different clients
with different baselines, where one fixed threshold would be too strict for
some and too loose for others.
